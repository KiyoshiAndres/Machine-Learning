```python
import matplotlib.pyplot as plt
import math
import numpy as np
import gudhi as gd
import matplotlib
```

We will create a circle with given radius as a discrete collection of points. A discrete collection of points does not usually have non-trivial first homology group, but with TDA we will get a non-trivial first homology group.

```python
radius = 1
number_of_points = 5
range_of_x = [radius * (x / number_of_points) for x in range(0, number_of_points + 1)]

circle_points = [ [x_sign * x, y_sign * math.sqrt(radius ** 2 - x ** 2)] for x in range_of_x for y_sign in [-1,1] for x_sign in [-1,1]]
```

```python
print(circle_points)
```

These lists are to plot the circle in pyplot, and unnecessary for the persistence homology computations.

```python
circle_x = [ x[0] for x in circle_points]
circle_y = [ x[1] for x in circle_points]
```

```python
# Scatter plot
plt.scatter(circle_x, circle_y)

plt.gca().set_aspect('equal')
plt.axis([-radius, radius, -radius, radius])

# Display the plot
plt.show()
```

### TDA

```python
alpha_complex_circle = gd.AlphaComplex(points=circle_points).create_simplex_tree()
persistence_of_circle = alpha_complex_circle.persistence()
```

```python
gd.plot_persistence_diagram(persistence_of_circle)
```

```python
gd.plot_persistence_barcode(persistence_of_circle)
```

## Figure 8

Shifting every point of the circle to the right by 2 $\times$ radius, and appending it to the original circle, we obtain a figure eight.

```python
figure_eight_points = [ [point[0] + shift, point[1]] for point in circle_points for shift in [0,2 * radius]]
```

```python
figure_eight_x = [ x[0] for x in figure_eight_points]
figure_eight_y = [ x[1] for x in figure_eight_points]
```

```python
# Scatter plot
plt.scatter(figure_eight_x, figure_eight_y)

plt.gca().set_aspect('equal')
plt.axis([-radius, 3 * radius, -radius, radius])

# Display the plot
plt.show()
```

```python
alpha_complex_figure_eight = gd.AlphaComplex(points=figure_eight_points).create_simplex_tree()
persistence_of_figure_eight = alpha_complex_figure_eight.persistence()
```

```python
gd.plot_persistence_diagram(persistence_of_figure_eight)
```
