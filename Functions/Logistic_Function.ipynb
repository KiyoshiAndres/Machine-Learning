From wikipedia, the logistic function is a sigmoid curve with equation
$$f(x)=\frac{L}{1+e^{-k(x-x_0)}}.$$
<br>
$k$ is the growth rate/steepness of the curve.
<br>
$x_0$ is the midpoint.
<br>
$L$ is the supremum.

```python
import matplotlib.pyplot as plt
import numpy as np

k = 1
x_0 = 0
L = 1

x_values = np.arange(-5,5,.1)
y_values = L / (1 + np.exp(-k * (x_values - x_0)))
```

```python
plt.plot(x_values, y_values)
plt.show()
```
