```python

# Creating the DataFrame
data = {
    'Year': [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Median house price': [340000, 370000, 350000, 420000, 425000, 500000, 520000, 550000, 596000, 610000, 660000, 720000, 730000, 760000],
    'Westpac: 4 year forecast': ['56%', '53%', None, '13%', '33%', '-5%', '45%', '34%', '34%', '20%', '-20%', '15%', '15%', '-10%'],
    'Joe Bloggs: 2 year forecast': ['23%', '34%', '19%', '42%', '23%', '15%', '15%', '18%', '19%', '23%', '13%', '8%', '7%', '-2%'],
    'Harry Spent: 5 year forecast': ['-20%', '-80%', '-70%', '-80%', '-50%', '-90%', '-30%', None, '-110%', '-90%', '-60%', '-69%', '-80%', '-80%']
}

df = pd.DataFrame(data)

# Converting percentage strings to numeric values
for column in df.columns[2:]:
    df[column] = df[column].str.replace('%', '').astype(float)

# Dropping the 'Median house price' column
df = df.drop(columns=['Median house price'])

# Plotting with matplotlib
plt.figure(figsize=(12, 8))
for column in df.columns[1:]:
    plt.plot(df['Year'], df[column], label=column)

plt.xlabel('Year')
plt.ylabel('Forecast Values (%)')
plt.title('Forecasts Over Years')
plt.legend()
plt.show()
```
